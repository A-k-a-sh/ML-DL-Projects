Here the genres are Action, Adventure, Fantasy.... we will make them as list in genres col like [Adventure, Fantasy...]

Here each genres are strings. if we remove the string , it will be list of obj

```python
import json
def convert(str_val , find_director : bool = False) -> list[str]:
    if isinstance(str_val, (list, dict)):
        return str_val
    parsed = json.loads(str_val)
    # print(parsed)
    make_list = []
    for i in parsed:
        if find_director:
            if i['job'] == 'Director':
                make_list.append(i['name'])
        else :
            make_list.append(i['name'])    
        
    return make_list

```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

Now each movie got a tag which is a string. we will convert each string to a vector. This is called vectorization or textvectorization.

Then if a user take a vector, we can recommend closer vector to the user.

There is many technique for converting text to vector:
* Bag of Words
* TF-IDF
* Count Vectorizer

more on - https://www.geeksforgeeks.org/nlp/vectorization-techniques-in-nlp/

---

Bag of words technique is we fist combine all the tags making a lagre text, finding most frequent words(how many? will be decided by us) from that large text. then for each movie tag, we will count the frequency of those words

---

but first, in each tage, there is similar words, like the words becomes and become is same. we will keep only the root word using a popular natural language toolkit : `nltk`

```python
import nltk
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
#ex of ps.stem

ps.stem('become'), ps.stem('becomes'),ps.stem('becoming') ##all produce same output
```

```python

def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
        
```

```python
#before
new_movies['tags'][0]
```

```python
##after

stem(new_movies['tags'][0])
```

```python
#apply on all the col
new_movies['tags'] = new_movies['tags'].apply(stem)
```

```python
#now performing vectorization using sklear
import sklearn 
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000 , stop_words = 'english')
```

Here the `max_features` is how many frequent words we take

and in tags, there is some english words eg: a, an , the, in , to which won't help us finding similarities.
they are `stop_words`

so we say remove those `stop_words` of english text

```python
vector = cv.fit_transform(new_movies['tags']).toarray()
vector.shape
```

```python
vector[0]
```

```python
#now finding similarites between vectors
from sklearn.metrics.pairwise import cosine_similarity
similarites = cosine_similarity(vector)
similarites.shape
```

```python
similarites[1] #if val is closer to 1, they are more similar
```

```python
def recommned(movie_name : str):
    movie_indx = new_movies.index[new_movies['title'].str.lower() == movie_name.lower()]
    if len(movie_indx) == 0:
        print('No movie found')
        return
    movie_indx = movie_indx[0]
    distances = sorted(list(enumerate(similarites[movie_indx])) , reverse = True, key = lambda x : x[1]) # a tuple
    for idx, dist in distances[1:11]:
        print(new_movies.iloc[idx]['title'])
        #or
        # print(new_movies['title'][idx])
```
