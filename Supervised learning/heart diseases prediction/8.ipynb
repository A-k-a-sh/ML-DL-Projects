### Feature importance

```python
from turtle import setup
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline


from importnb import imports
with imports("ipynb"):
    import setUp 
df = setUp.df    
X = setUp.X
y = setUp.y
X_train , X_test , y_train , y_test = setUp.X_train , setUp.X_test , setUp.y_train , setUp.y_test

```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(C=0.20433597178569418, solver='liblinear') #found the best estimator using GridSearchCV

clf.fit(X_train , y_train)

clf.score(X_test , y_test)
```

```python
clf.coef_
```

```python
df.columns
```

```python
feature_dict = dict(zip(df.columns , clf.coef_[0]))

feature_dict
```

```python
feature_Df = pd.DataFrame(feature_dict , index=['Vals'])

feature_Df
```

```python
feature_Df.T.index
```

```python
feature_Df.T.plot.bar()
```
