# Run from kaggle

To run from kaggle, you first get he public link of you file. 

Copy the File ID (the long string in the URL like https://drive.google.com/file/d/FILE_ID/view).


In `output_file` give the file name and run the cell. This will download your file from your drive kaggle output dir : `/kaggle/working`

```python
import gdown

# Replace with your files ID
file_id = "1_8aVuLDjQu6_L6Axbz8hNW46uu6ulenC"
url = f"https://drive.google.com/uc?id={file_id}"

output_file = "12. Setting Up an Optimizer and a Loss Function.mp4"
gdown.download(url, output_file, quiet=False)
```

```python
input_file = output_file
output_dir = "/kaggle/working"

!whisper "{input_file}" --model small --output_format srt --output_dir "{output_dir}"
```
